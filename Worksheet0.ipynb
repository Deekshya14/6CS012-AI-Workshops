**Task 1 — Unit Conversion Program**

```python
def unit_converter():
    """
    Converts units for length, weight, and volume.

    Parameters:
    None (takes user input)

    Returns:
    float : converted value
    """

    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter choice (1/2/3): ")

        value = float(input("Enter the value to convert: "))

        if choice == "1":
            print("1. meters to feet")
            print("2. feet to meters")

            sub = input("Choose option: ")

            if sub == "1":
                result = value * 3.28084
                print("Result:", result, "ft")

            elif sub == "2":
                result = value / 3.28084
                print("Result:", result, "m")

        elif choice == "2":
            print("1. kg to pounds")
            print("2. pounds to kg")

            sub = input("Choose option: ")

            if sub == "1":
                result = value * 2.20462
                print("Result:", result, "lbs")

            elif sub == "2":
                result = value / 2.20462
                print("Result:", result, "kg")

        elif choice == "3":
            print("1. liters to gallons")
            print("2. gallons to liters")

            sub = input("Choose option: ")

            if sub == "1":
                result = value * 0.264172
                print("Result:", result, "gal")

            elif sub == "2":
                result = value / 0.264172
                print("Result:", result, "L")

        else:
            print("Invalid conversion type")

    except ValueError:
        print("Invalid input. Please enter numeric values.")

unit_converter()
```

**Task 2 — Mathematical Operations on List**

```python
def calculate_sum(numbers):
    """Returns the sum of a list of numbers."""
    return sum(numbers)

def calculate_average(numbers):
    """Returns the average of a list of numbers."""
    return sum(numbers) / len(numbers)

def find_max(numbers):
    """Returns the maximum value."""
    return max(numbers)

def find_min(numbers):
    """Returns the minimum value."""
    return min(numbers)

try:
    data = input("Enter numbers separated by spaces: ")
    numbers = list(map(float, data.split()))

    print("Choose operation:")
    print("1 Sum")
    print("2 Average")
    print("3 Maximum")
    print("4 Minimum")

    choice = input("Enter choice: ")

    if choice == "1":
        print("Sum:", calculate_sum(numbers))

    elif choice == "2":
        print("Average:", calculate_average(numbers))

    elif choice == "3":
        print("Maximum:", find_max(numbers))

    elif choice == "4":
        print("Minimum:", find_min(numbers))

    else:
        print("Invalid operation")

except ValueError:
    print("Invalid input. Please enter numbers.")
```

# **Exercise on List Manipulation:**

**1 Extract every other element**

```python
def extract_every_other(lst):
    """Returns every other element from the list."""
    return lst[::2]

print(extract_every_other([1,2,3,4,5,6]))
```

**2 Get sublist**

```python
def get_sublist(lst, start, end):
    """Returns sublist from start to end index inclusive."""
    return lst[start:end+1]

print(get_sublist([1,2,3,4,5,6],2,4))
```

**3 Reverse list**

```python
def reverse_list(lst):
    """Returns reversed list using slicing."""
    return lst[::-1]

print(reverse_list([1,2,3,4,5]))
```

**4 Remove first and last**

```python
def remove_first_last(lst):
    """Removes first and last element."""
    return lst[1:-1]

print(remove_first_last([1,2,3,4,5]))
```

**5 Get first n**

```python
def get_first_n(lst, n):
    """Returns first n elements."""
    return lst[:n]

print(get_first_n([1,2,3,4,5],3))
```

**6 Get last n**

```python
def get_last_n(lst, n):
    """Returns last n elements."""
    return lst[-n:]

print(get_last_n([1,2,3,4,5],2))
```

**7 Reverse skip**

```python
def reverse_skip(lst):
    """Returns every second element from second last in reverse."""
    return lst[-2::-2]

print(reverse_skip([1,2,3,4,5,6]))
```

# **Exercise on Nested List:**

**1 Flatten list**

```python
def flatten(lst):
    """Flattens one level nested list."""
    result = []
    for sub in lst:
        result.extend(sub)
    return result

print(flatten([[1,2],[3,4],[5]]))
```

**2 Access nested element**

```python
def access_nested_element(lst, indices):
    """Access element using list of indices."""
    element = lst
    for i in indices:
        element = element[i]
    return element

lst = [[1,2,3],[4,5,6],[7,8,9]]
print(access_nested_element(lst,[1,2]))
```

**3 Sum nested**

```python
def sum_nested(lst):
    """Returns sum of all elements in nested list."""
    total = 0
    for item in lst:
        if isinstance(item,list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

**4 Remove element**

```python
def remove_element(lst, elem):
    """Removes element from nested list."""
    result = []
    for item in lst:
        if isinstance(item,list):
            result.append(remove_element(item,elem))
        elif item != elem:
            result.append(item)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

**5 Find max**

```python
def find_max(lst):
    """Finds maximum in nested list."""
    maximum = float('-inf')
    for item in lst:
        if isinstance(item,list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum,item)
    return maximum

print(find_max([[1,2],[3,[4,5]],6]))
```

**6 Count occurrences**

```python
def count_occurrences(lst, elem):
    """Counts occurrences in nested list."""
    count = 0
    for item in lst:
        if isinstance(item,list):
            count += count_occurrences(item,elem)
        elif item == elem:
            count += 1
    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

**7 Deep flatten**

```python
def deep_flatten(lst):
    """Flattens list of any depth."""
    result = []
    for item in lst:
        if isinstance(item,list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

**8 Nested average**

```python
def average_nested(lst):
    """Returns average of nested list."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```

**Problem 4 — Matrix Operations**

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])

# 1 Prove A * A^-1 = I
A_inv = np.linalg.inv(A)

result = np.dot(A, A_inv)

print("A * A^-1 =\n", result)

# 2 Prove AB ≠ BA
AB = np.dot(A,B)
BA = np.dot(B,A)

print("\nAB =\n", AB)
print("\nBA =\n", BA)

# 3 Prove (AB)^T = B^T A^T

left = (AB).T
right = np.dot(B.T, A.T)

print("\n(AB)^T =\n", left)
print("\nB^T A^T =\n", right)
```
